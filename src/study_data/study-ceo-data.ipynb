```python
# check whether for each REO the SECCIO_REAL contain a subset of SECCIO_REAL in the previous year
# Group by SECCIO_REAL and count unique REOs for each SECCIO_REAL
seccio_reo_count = df.groupby(["COMARCA", "MUNICIPI", "SECCIO_REAL"])["REO"].nunique()

# Filter SECCIO_REAL with more than one unique REO
seccio_multiples = seccio_reo_count[seccio_reo_count > 1].index.tolist()

seccio_reo_count.sort_values(ascending=False)

len(seccio_multiples)
```
